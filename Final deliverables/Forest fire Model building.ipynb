importing Required libraries

```python
import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
%cd ../test_set #Test set navigation
```

Define parameters and arguments for ImageDataGenerator

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                 shear_range=0.2,
                                 rotation_range=180,
                                 zoom_range=0.2,
                                 horizontal_flip=True)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
import tensorflow as tf
```

Applying ImageDataGenerator

```python
x_test = test_datagen.flow_from_directory(r'/content/drive/MyDrive/Dataset/Dataset/test_set',
                                          target_size=(128,128),
                                          batch_size=32,
                                          class_mode='binary')
```

```python
x_train = train_datagen.flow_from_directory(r'/content/drive/MyDrive/Dataset/Dataset/train_set',
                                          target_size=(128,128),
                                          batch_size=32,
                                          class_mode='binary')
```

Model building

```python
#import model building libraries

#to define linear initialisation import sequential
from keras.models import Sequential
#to add layer import Dense
from keras.layers import Dense
#to create convolution kernel import convolution2D
from keras.layers import Convolution2D
#import Maxpooling layer
from keras.layers import MaxPool2D
#import flatten layer
from keras.layers import Flatten
import warnings
warnings.filterwarnings('ignore')
```

```python
#initialize the model
model=Sequential()
```

```python
from keras.layers.pooling.max_pooling2d import MaxPooling2D
#add convolution layer
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation='relu'))
#add maxpooling layer
model.add(MaxPooling2D(pool_size=(2,2)))
#add flatten layer
model.add(Flatten())
```

Adding dense layer

```python
#add hidden layer
model.add(Dense(150,activation='relu'))
#add output layer
model.add(Dense(1,activation='sigmoid'))
```

```python
#configure the learning process
model.compile(loss = 'binary_crossentropy',
              optimizer = "adam",
              metrics = ["accuracy"])
```

```python
#training the model
model.fit(x_train,steps_per_epoch=14,
                    epochs=10,validation_data=x_test,
                    validation_steps=4)
```

```python
#save the model
model.save("forest1.h5")
```
